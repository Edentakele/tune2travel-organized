```python
import pandas as pd

# Load the dataset into a Pandas DataFrame
df = pd.read_csv("./prod/despacito_en_api.csv")

# Show first few rows
df.head()
```

```python
# Import Hugging Face pipeline
from transformers import pipeline

# Load the sentiment analysis model
sentiment_model = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
from tqdm import tqdm

# Ensure progress bar appears
tqdm.pandas()

# Function to process sentiment safely
def get_sentiment(text):
    text = str(text)[:512]  # Ensure text fits within BERT's token limit
    result = sentiment_model(text)[0]  # Get sentiment
    return result['label']  # Returns 'POSITIVE' or 'NEGATIVE'

# Apply sentiment analysis with a progress bar
df['sentiment'] = df['comment'].progress_apply(get_sentiment)

# Show results
df[['comment', 'sentiment']].head()

print(df.columns)  # List all column names

df.rename(columns={df.columns[1]: "comment"}, inplace=True)  # Rename the second column to 'comment'
```

```python
from transformers import pipeline
from tqdm import tqdm

# Load Hugging Face sentiment analysis model
sentiment_model = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# Enable progress bar
tqdm.pandas()

# Function to process sentiment safely
def get_sentiment(text):
    text = str(text)[:512]  # Ensure text fits within BERT's token limit
    result = sentiment_model(text)[0]  # Get sentiment
    return result['label']  # Returns 'POSITIVE' or 'NEGATIVE'

# Apply sentiment analysis to comments
df['sentiment'] = df['comment'].progress_apply(get_sentiment)

# Show results
df[['comment', 'sentiment']].head()
```

```python
import matplotlib.pyplot as plt

# Count sentiment categories
sentiment_counts = df['sentiment'].value_counts()

# Plot results
plt.figure(figsize=(6,4))
sentiment_counts.plot(kind='bar', color=['green', 'red'])
plt.title("Sentiment Analysis of YouTube Comments (BERT)")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

df.to_csv("despactio_bert_sentiment_results.csv", index=False)

# Provide download link
from google.colab import files
files.download("bert_sentiment_results.csv")
```
